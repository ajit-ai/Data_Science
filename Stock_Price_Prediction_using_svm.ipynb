```python
# Machine learning
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# For data manipulation
import pandas as pd
import numpy as np

# To plot
import matplotlib.pyplot as plt
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Read the csv file using read_csv 
# method of pandas
df = pd.read_csv('content/RELIANCE.csv')
df
```

```python
# Changes The Date column as index columns
df.index = pd.to_datetime(df['Date'])
df

# drop The original date column
df = df.drop(['Date'], axis='columns')
df
```

```python
# Create predictor variables
df['Open-Close'] = df.Open - df.Close
df['High-Low'] = df.High - df.Low

# Store all predictor variables in a variable X
X = df[['Open-Close', 'High-Low']]
X.head()
```

```python
# Target variables
y = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
y
```

```python
split_percentage = 0.8
split = int(split_percentage*len(df))

# Train data set
X_train = X[:split]
y_train = y[:split]

# Test data set
X_test = X[split:]
y_test = y[split:]
```

```python
# Support vector classifier
cls = SVC().fit(X_train, y_train)
```

```python
df['Predicted_Signal'] = cls.predict(X)
```

```python
df['Predicted_Signal'] = cls.predict(X)
```

```python
# Calculate daily returns
df['Return'] = df.Close.pct_change()
```

```python
# Calculate strategy returns
df['Strategy_Return'] = df.Return *df.Predicted_Signal.shift(1)
```

```python
# Calculate Cumulutive returns
df['Cum_Ret'] = df['Return'].cumsum()
df
```

```python
# Plot Strategy Cumulative returns
df['Cum_Strategy'] = df['Strategy_Return'].cumsum()
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(df['Cum_Ret'],color='red')
plt.plot(df['Cum_Strategy'],color='blue')
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score

y_pred = cls.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"ROC AUC: {roc_auc}")
print(f"F1 Score: {f1}")
```

```python
print("The data was split into training and testing sets using an 80/20 split. 80% of the data was used for training, and the remaining 20% was used for testing.")

# Calculate training accuracy
train_accuracy = accuracy_score(y_train, cls.predict(X_train))

# Calculate testing accuracy
test_accuracy = accuracy_score(y_test, cls.predict(X_test))

print(f"Training Accuracy: {train_accuracy}")
print(f"Testing Accuracy: {test_accuracy}")
```


Create and train SVC models with different kernels and calculate their accuracy on the test set.



```python
from sklearn.metrics import accuracy_score

# Linear kernel
cls_linear = SVC(kernel='linear').fit(X_train, y_train)
y_pred_linear = cls_linear.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)
print(f"Accuracy with Linear Kernel: {accuracy_linear}")

# Polynomial kernel
cls_poly = SVC(kernel='poly', degree=3).fit(X_train, y_train)
y_pred_poly = cls_poly.predict(X_test)
accuracy_poly = accuracy_score(y_test, y_pred_poly)
print(f"Accuracy with Polynomial Kernel (degree=3): {accuracy_poly}")

# RBF kernel (default)
cls_rbf = SVC(kernel='rbf').fit(X_train, y_train)
y_pred_rbf = cls_rbf.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
print(f"Accuracy with RBF Kernel: {accuracy_rbf}")

# Sigmoid kernel
cls_sigmoid = SVC(kernel='sigmoid').fit(X_train, y_train)
y_pred_sigmoid = cls_sigmoid.predict(X_test)
accuracy_sigmoid = accuracy_score(y_test, y_pred_sigmoid)
print(f"Accuracy with Sigmoid Kernel: {accuracy_sigmoid}")
```

## Summary:

### Data Analysis Key Findings

*   The initial SVM model on the test set achieved an accuracy of approximately 0.472, a ROC AUC of approximately 0.472, and an F1 Score of approximately 0.502.
*   The data was split into training and testing sets using an 80/20 ratio.
*   The training accuracy of the initial model was approximately 0.536, while the testing accuracy was approximately 0.472.
*   Exploring different kernels for the SVC model resulted in varying accuracies on the test set: Linear (0.491), Polynomial (0.490), RBF (0.472), and Sigmoid (0.505).

### Insights or Next Steps

*   The relatively low accuracy across all tested kernels suggests that the current features or the SVM model itself may not be sufficiently powerful for this stock price prediction task. Further feature engineering or exploration of more complex models might be beneficial.
*   Hyperparameter tuning for the different kernels, especially for the polynomial and RBF kernels which have additional parameters (degree, gamma, C), could potentially improve model performance.
