```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data = pd.read_csv("sample_data/general.csv")
data.head()
```

```python
# @title BALANCE_FREQUENCY

from matplotlib import pyplot as plt
data['BALANCE_FREQUENCY'].plot(kind='line', figsize=(8, 4), title='BALANCE_FREQUENCY')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title BALANCE_FREQUENCY

from matplotlib import pyplot as plt
data['BALANCE_FREQUENCY'].plot(kind='hist', bins=20, title='BALANCE_FREQUENCY')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
